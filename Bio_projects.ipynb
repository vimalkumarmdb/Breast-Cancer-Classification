```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Download the dataset from UCI reposetry

```
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
cancer.keys()
```

```
print(cancer['DESCR'])
```

```
print(cancer['target_names'])
```

```
print(cancer['target'])
```

```
print(cancer['feature_names'])
```

```
cancer['data'].shape
```

```
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```
df.head()
```

```
df.tail()
```

## Visulization of the some feature V/S target

```
import seaborn as sn
```

```
sn.pairplot (df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
plt.show()
```

```
sn.countplot(df['target'], label = "Count") 
plt.show()
```

```
sn.scatterplot(data = df,x = 'mean area', y = 'mean smoothness', hue = 'target')
plt.show()
```

```
import pandas_profiling 
pfr = pandas_profiling.ProfileReport(df,correlation_threshold = 0.5,
                                     bins = 20)
pfr.to_file("Analysis_report.html")
```

```
X = df.drop(['target'],axis=1)
```

```
X.head()
```

```
y = df['target']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix
svc_model = SVC( kernel='rbf',gamma='auto')
svc_model.fit(X_train, y_train)
```

```
predict_y = svc_model.predict(X_test)
cm = confusion_matrix(y_test, predict_y)
```

```
#!pip install seaborn 
```

```
#%matplotlib inline
plt.figure(figsize=(6,5))
sn.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["No", "Yes"] , yticklabels = ["No", "Yes"],)
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
plt.show()
```

```
print(classification_report(y_test, predict_y))
```

# Let's improve the model

*   Scaled the x_train and X_test data




```
min_train = X_train.min()
min_train
```

```
range_train = (X_train - min_train).max()
range_train
```

```
X_train_scaled = (X_train - min_train)/range_train
```

```
sn.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
plt.show()
```

```
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

* For parameter optimization used Grid Search algorithms 

```
#param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
#from sklearn.model_selection import GridSearchCV
#grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
#grid.fit(X_train_scaled,y_train)
svc_model = SVC(C=10, kernel='rbf',gamma=1)
svc_model.fit(X_train_scaled,y_train)
```

```
#grid.best_params_
```

```
#grid.best_estimator_
```

```
svm1_predictions = svc_model.predict(X_test_scaled)
```

```
CM = confusion_matrix(y_test, svm1_predictions)
```

```
#%matplotlib inline
plt.figure(figsize=(5,5))
sn.heatmap(CM, annot=True,  fmt='d', xticklabels = ["No", "Yes"] , yticklabels = ["No", "Yes"],)
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
plt.show()
```

```
from sklearn.metrics import accuracy_score,average_precision_score,f1_score
```

```
print("classification report:\n",classification_report(y_test, svm1_predictions))
```

```
print("Accuracy score: {0:.2f}".format(accuracy_score(y_test, svm1_predictions)*100))
```

```
print("Average precision score:{0:.2f}".format(average_precision_score(y_test, svm1_predictions)*100))
```

```
print("f1_score:{0:.2f}".format(f1_score(y_test,svm1_predictions)*100))
```

## ====================== Knn Model ======================

```
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=12)
neigh.fit(X_train_scaled, y_train)
```

```
y_pred_knn = neigh.predict(X_test_scaled)
```

```
cm1 = confusion_matrix(y_test,y_pred_knn)
```

```
plt.figure(figsize=(6,5))
sn.heatmap(cm1, annot=True,  fmt='d', xticklabels = ["No", "Yes"] , yticklabels = ["No", "Yes"])
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
plt.show()
```

```
print("classification report:\n",classification_report(y_test, y_pred_knn))
```

```
print("Accuracy score: {0:.2f}".format(accuracy_score(y_test,y_pred_knn)*100))
```

```
print("Average precision score:{0:.2f}".format(average_precision_score(y_test,y_pred_knn)*100))
```

```
print("f1_score:{0:.2f}".format(f1_score(y_test,y_pred_knn)*100))
```

## ============= SVM-Knn-Ensemble(Voting)============

---



```
from sklearn.ensemble import  VotingClassifier
```

```
Svc_Model = SVC(C=3, kernel='rbf',gamma='auto')
```

```
eclf1 = VotingClassifier(estimators=[('SVM', Svc_Model), ('Knn', neigh), ], voting='hard')
eclf1 = eclf1.fit(X_train_scaled, y_train)
```

```
y_pred_Enseble =eclf1.predict(X_test_scaled)
```

```
cm3 = confusion_matrix(y_test,y_pred_Enseble)
```

```
cm3
```

```
plt.figure(figsize=(5,4))
sn.heatmap(cm3, annot=True,  fmt='d', xticklabels = ["No", "Yes"] , yticklabels = ["No", "Yes"])
plt.ylabel('True label',fontsize=12)
plt.xlabel('Predicted label',fontsize=12)
plt.show()
```

```
print("classification report:\n",classification_report(y_test,y_pred_Enseble))
```

```
print("Accuracy score: {0:.2f}".format(accuracy_score(y_test,y_pred_Enseble)*100))
```

```
print("Average precision score:{0:.2f}".format(average_precision_score(y_test,y_pred_Enseble)*100))
```

```
print("f1_score:{0:.2f}".format(f1_score(y_test,y_pred_Enseble)*100))
```
